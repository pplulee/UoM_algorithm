```python
from hashset import hashset
import random
import time
import pandas as pd

test_mode=4

def random_string(size=10):
    return ''.join(random.sample(['z','y','x','w','v','u','t','s','r','q','p','o','n','m','l','k','j','i','h','g','f','e','d','c','b','a'], size))
result_insert = []
result_find = []
df =pd.DataFrame(columns=["size","avg_insert","max_insert","avg_find","max_find","collision_rate","time"])
```

```python
for i in range(100,100000,100):
    print(i)
    words = []
    for j in range(0,i):
        words.append(random_string())
    test_set = hashset(i*2)
    start = time.time()
    for word in words:
        test_set.insert(word)
    for word in words:
        test_set.find(word)
    end = time.time()
    df.loc[df.index.size]=[i,test_set.get_avg_insert(),test_set.get_max_insert(),test_set.get_avg_find(),test_set.get_max_find(),test_set.get_collision()/test_set.hash_table_size,end-start]
    del words
    del test_set
```
