```python
from bstree import bstree
import random
import time
import pandas as pd


def random_string(size=10):
    return ''.join(random.sample(['z','y','x','w','v','u','t','s','r','q','p','o','n','m','l','k','j','i','h','g','f','e','d','c','b','a'], size))
result_insert = []
result_find = []
df =pd.DataFrame(columns=["size","height","avg_insert","avg_find","time"])
```

```python
for i in range(100,100000,100):
    words = []
    for j in range(0,i):
        words.append(random_string())
    test_set = bstree()
    start = time.time()
    for word in words:
        test_set.insert(word)
    for word in words:
        test_set.find(word)
    end = time.time()
    print(i)
    df.loc[df.index.size]=[test_set.size(),test_set.height(),test_set.get_insert_stats()/i,test_set.get_find_stats()/i,end-start]
    del words
    del test_set
```

```python
print(len(df))
df.to_csv("bstree_result.csv")
df
```
